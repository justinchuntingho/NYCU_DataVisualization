## Reading .txt

```python
wordlist = [] # Creating an empty list

with open('Example_word_list.txt', mode = 'r', encoding='utf-8') as f:
    for line in f:
        wordlist.append(line) #Appending to the empty list
```

```python
wordlist
```

```python
wordlist_clean = [x.strip() for x in wordlist] # This is how to clean the "\n", you can ust add it in the loop.
wordlist_clean
```

```python
#Changing txt - adding a word
wordlist_clean.append('new_word')
```

```python
wordlist_clean
```

```python
# Writing to newfile
with open('newwordlist.txt', mode='w', encoding='utf-8') as f: # Note the mode='w'
    for word in wordlist_clean:
        f.write(word + '\n')
```

#### What if I only want data from the stops at NYCU?

We know `item['場站名稱']` can get us the station name.

We also know `if "XXXX" in "Some Texts Blah Blah Blah"` can tells us if the text contains XXXX

We know how to filter a list

```python
filtered = []
for item in data:
    if '陽明交通大學' in item['場站名稱']:
        filtered.append(item)
filtered
```

```python
# This is how you write a .json file
with open('filtered.json', 'w', encoding='utf-8') as f:
    json.dump(filtered, f)
```

## Writing and loading files with pickle

pickle is a way to store file as is

```python
import pickle 

with open('filtered.pkl', 'wb') as handle:
    pickle.dump(filtered, handle)
```

```python
# Load data (deserialize)
with open('filtered.pkl', 'rb') as handle:
    my_data = pickle.load(handle)

my_data
```
