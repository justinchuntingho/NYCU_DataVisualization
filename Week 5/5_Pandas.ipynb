## Functions

````python
def add_one(x):
    x = int(x)
    new_number = x+1
    return new_number
````

In the **first line**:
* ```def``` indicates to Python that you are creating (defining) a function
* ```add_one``` is the name of the function. You can give it any name you want, as long as (a) it does not start with a number or restricted character, (b) it does not have a space, and (c) it does not overlap with one of Python's basic commands.
* ```(x)``` specifies the arguments that will be passed along to the function (i.e., that the function will use). It is up to you what name you give to the argument. A function can also be written without arguments - i.e., just ```()``` - but in this course we mostly will use arguments.
* ```:``` don't forget this - as it indicates that the function has been defined, and what comes underneath is the function itself

**After the first line**, notice the indent, i.e., that anything that happens inside the function is indented (with spaces) within the function. That way, Python knows until which line of the code belongs to the function.

In this area, the function is being defined. The arguments - in our case, ```x``` - can then be used by the function in whatever operations we want to run. In this case, we are saying to add 1 to x or, in code, ```x = x + 1```.

**In the last line of the function**, we have ```return```, which specifies what the function should return.

````python
add_one(1)
````

#### Let's fix the mess in usage data...

Recap: Most of the usage are integers, but for ones that stopped working sometimes it is noted as `"停運"`, sometimes it is just `""`. 

Remember this is real world data, and in the real world, people do not always follow the best data practices.

````python
# First getting the names of all the columns that has usage data:
day_cols = [col for col in df.columns if "使用筆數" in col]
day_cols
````

````python
# pd.to_numeric() is a function to change things to numbers
pd.to_numeric([1, 2, '3', 'Nothing'], errors="coerce")
````

````python
# Why we can't use int()?
int([1, 2, '3', 'Nothing'])
````

````python
# Also can't do this
[int(x) for x in [1, 2, '3', 'Nothing']]
````

Conclusion: pandas functions are much more powerful

````python
# Apply to one column
df["日期29日使用筆數"] = pd.to_numeric(df["日期29日使用筆數"], errors="coerce")
````

````python
# We can use for loop too
for col in day_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")
````

````python
# Even better, we can use apply()
# df[day_cols] is telling pandas, let's extract only the usage columns
# df.apply() is a function to take a function and apply it to everything column/row, one by one
df[day_cols] = df[day_cols].apply(pd.to_numeric, errors="coerce")
````

````python
# Let's have a look, now everything is numbers, finally
df.dtypes
````

### Min, Max, Sum, and more...

| Function    | Description              | Example                  |
| ----------- | ------------------------ | ------------------------ |
| `sum()`     | Total sum of values      | `df[day_cols].sum()`     |
| `mean()`    | Average value            | `df[day_cols].mean()`    |
| `max()`     | Maximum value            | `df[day_cols].max()`     |
| `min()`     | Minimum value            | `df[day_cols].min()`     |
| `median()`  | Middle value             | `df[day_cols].median()`  |
| `std()`     | Standard deviation       | `df[day_cols].std()`     |
| `count()`   | Count of non-null values | `df[day_cols].count()`   |
| `nunique()` | Number of unique values  | `df[day_cols].nunique()` |
| `idxmax()`  | Index of max value       | `df[day_cols].idxmax()`  |
| `idxmin()`  | Index of min value       | `df[day_cols].idxmin()`  |

**Important idea**
- `axis=1` = operating across columns, do it row by row
- `axis=0` = operating across rows, do it column by column

````python
df[day_cols].mean(axis=1) # axis=1 tell pandas to run it by rows, axis=0 will run it by columns
````

````python
# This is how you put the results back as a new column
df["daily_avg"] = df[day_cols].mean(axis=1)
````

````python
# We can use the max() metho to find the busiest day per station
df["max_daily_usage"] = df[day_cols].max(axis=1)
````

````python
df["total_usage"] = df[day_cols].sum(axis=1)
````
