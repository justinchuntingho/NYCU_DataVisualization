## Working with Dictionaries

```python
# Create a list of emails
# For loops for dictionaries 
emails = {"Justin":"jcho@nycu.edu.tw",
         "Peter":"happypancake@gmail.com"}

for key,value in emails.items():
    print(key, "has email", value)

# Note remember you need to use .items() and TWO iterative variables to catch them
```

```python
# If you only want values
emails.values()
```

```python
# Only keys
emails.keys()
```

```python
#You can also iterate only over the keys or only over the values 
for hello_kitty in emails.keys():
    print(hello_kitty)
```

# Conditional statements (in loops)

```python
# Print number smaller than 5
# Operation on every element
my_list = [3,4,5,6]

newnumbers=[]

for number in my_list:
    if number < 5:
        print(number)
        newnumbers.append(number)
    else:
        print("larger than 5")
```

```python
print(newnumbers)
```

```python
# While loop
counter = 0
while counter <= 4:
    print("Looping...")
    print(counter)
    counter += 1
else:
    print("Else!")
```

```python
#create count how many times social media mentioned
socialmedia = ["Tiktok", "Facebook", "Instagram", "Snapchat", "Twitter", "LinkedIn", "Tiktok", "Instagram", "Facebook", "Tiktok"]

counter = 0
socialmedialist = ["Tiktok", "Instagram", "Snapchat"]

for medium in socialmedia:
    if medium in socialmedialist:
        print("Yes")
        counter += 1
        print(counter)
    else:
        pass # pass is useless, just a placeholder here

```
